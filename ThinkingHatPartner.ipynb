```python
df = pd.read_table('hatPrompts.txt', index_col='Hat')
df
```

Function to get the response to a hat query from OpenAI

```python
def getHatResponse(hat, idea, role, content):
    print('Checking with the ' + hat + ' hat...\n')
    response = openai.ChatCompletion.create(
      model="gpt-3.5-turbo",
      messages=[
            {"role": "system", "content": role},
            {"role": "user", "content": idea},
            {"role": "user", "content": content}      ]
    )

    message = response['choices'][0]['message']['content']
    return message
```

Propose your idea and iterate through the different hats. You can remove the "sleep" statement if you are paying for faster access.

```python
idea = "I'm thinking of starting a t-shirt business at the Fairfield beach in Connecticut."
results = ''
for hat, row in df.iterrows():
    if hat!='Blue':
        msg = getHatResponse(hat, idea, row['Role'], row['Content'])
        results += msg
    else:
        msg = getHatResponse(hat, results, row['Role'], row['Content'])
    
    print(msg)
    print('\n-----------------------------------------------\n')
    sleep(3)
```
